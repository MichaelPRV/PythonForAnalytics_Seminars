1.2 Создать новый признак delta_renovated, который будет содержать разницу в годах между годом реновации и годом постройки дома. Если реновации дома не было, то в новом признаке поставьте "0"

```python
def renovation_time(ren_year, built_year):
    if ren_year:
        return ren_year - built_year
    else:
        return ren_year


df['delta_renovated'] = df.apply(lambda x: renovation_time(x['yr_renovated'], x['yr_built']), axis=1)
df.head(2)
```

1.3 Создайте признаки года продажи, месяца продажи

```python
df['year_of_sale'] = df['date'].apply(lambda x: int(x[:4]))
df.sample()
```

```python
df['month_of_sale'] = df['date'].apply(lambda x: int(x[4:6]))
df.sample()
```

1.4 Удалите признаки date, zipcode, lat, long

```python
df.drop(columns=['date', 'zipcode', 'lat', 'long'], inplace=True)
df.head(2)
```

3.1 Найдите среднюю стоимость домов в зависимости от количества спален.
Отсортируйте от меньшей стоимости к большей

```python
grouped_df = df.groupby('bedrooms')
grouped_df.agg({'price': 'mean'}).sort_values('price').round()
```
